**Creating Arrays**

```python
# creating a 1 dim array from Python lists or tuples:
a = np.array([1,2,3,4])
print(a)

b = np.array([[1, 2, 3,4,5], [4, 5, 6,7,8]])  # 2D array

#Creating arrays filled with zeros or ones:
zeros = np.zeros(5)  # Creates an array of 5 zeros
ones = np.ones((2, 3))  # Creates a 2x3 array of ones
z = np.full((5,4),5) # array of a particular value
I = np.eye(5) #identity matrix
Random = np.random.random((3,4)) #array with random values
range_array = np.random.randint(10,100,(3,5)) #random integer values array within a specific range
evenspace_value = np.linspace(10,30,5) #array of evenly spaced values --> specifying the number of values required
evenspace_step = np.arange(10,30,5) # array of evenly spaced values --> specifying the step

#Creating sequences of numbers:
sequence = np.arange(1, 11)  # Creates an array from 1 to 10
```

```python
print(b.shape)  #dimension

print(b.dtype)#Datatype

print(b.ndim)   #No. of dimensions

print(b.size)   #No. of elements
```

```python
# convert a list to a numpy array
list2 = [10,20,20,20,50]

np_array = np.asarray(list2)
print(np_array)
type(np_array)
```

**Array operations**

```python
arr1=np.array([(1,2,3,4),(5,6,7,8)])
arr2=np.array([(10,20,50,60),(50,60,10,70)])

result_add = arr1 + arr2  # Addition
result_mul = arr1 * 2  # Scalar multiplication

print(result_add)

print(result_mul)
```

```python
print(sum(arr1))
print(arr1.mean())
print(arr1.max())
print(arr1.min())
```
