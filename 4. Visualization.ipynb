# **Matplotlib**

```python

# importing matplotlib library
import matplotlib.pyplot as plt
```

```python

# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#Plotting the data
```

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python

# cosine wave
plt.plot(x,z)
plt.show()
```

```python

# adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python

# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python

plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
#Bar Plot
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#Pie Chart
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.2f%%')
plt.show()
```

```python
#scatter plot
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D Scatter Plot
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```
