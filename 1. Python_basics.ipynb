## **String**

Features:
1. It can contain letters, numbers, and special characters.
2. It can be enclosed in single ('') or double ("") quotes.
3. It can be indexed and sliced to access specific characters or substrings.
4. It is immutable, meaning you cannot change individual characters once defined.
5. It can be concatenated with other strings using the + operator.
6. It can be formatted using string interpolation methods like f-strings (in Python 3.6+).
7. It can be iterated through using loops.
8. It can have various string methods applied to it (e.g., upper(), lower(), split(), find()).
9. It can contain escape sequences (e.g., \n for newline, \t for tab).
10. It can be compared with other strings for equality or ordering.



```python
my_string = "Hello, Python!"

# Features and Operations:
# 1. String Creation:
#    - You can create a string by enclosing text in single (' ') or double (" ") quotes.

# 2. Length of String:
string_length = len(my_string)
# Output: 13 (counting each character, including spaces and punctuation)

# 3. Indexing and Slicing:
first_character = my_string[0]
second_to_fifth_characters = my_string[1:5]
last_character = my_string[-1]
# Output: 'H', 'ello', '!'

# 4. Concatenation:
new_string = my_string + " Welcome to Python!"
# Output: 'Hello, Python! Welcome to Python!'

# 5. Repetition:
repeated_string = my_string * 3
# Output: 'Hello, Python!Hello, Python!Hello, Python!'

# 6. String Methods:
uppercase_string = my_string.upper()
lowercase_string = my_string.lower()
split_string = my_string.split(', ')
# Output: 'HELLO, PYTHON!', 'hello, python!', ['Hello', 'Python!']

# 7. String Interpolation (f-strings, Python 3.6+):
name = "Alice"
age = 30
formatted_string = f"My name is {name} and I am {age} years old."
# Output: 'My name is Alice and I am 30 years old.'

# 8. Escape Sequences:
escaped_string = "This is a new line.\nThis is a tab\tThis is a backslash: \\"
# Output:
# 'This is a new line.'
# 'This is a tab  This is a backslash: \'

# 9. Checking for Substrings:
contains_python = "Python" in my_string
# Output: True

# 10. String Length:
string_length = len(my_string)
# Output: 13

# 11. String Comparison:
compare_strings = my_string == "Hello, Python!"
# Output: True

# 12. String Trimming:
trimmed_string = my_string.strip()
# Output: 'Hello, Python!'

# 13. Replacing Substrings:
replaced_string = my_string.replace("Python", "World")
# Output: 'Hello, World!'

# 14. Checking for Alphanumeric Characters:
is_alphanumeric = my_string.isalnum()
# Output: False

# 15. Finding Substrings:
position = my_string.find("Python")
# Output: 7 (position of "Python" in the string)

# 16. Checking for Uppercase and Lowercase:
is_upper = my_string.isupper()
is_lower = my_string.islower()
# Output: False (for both)

# 17. Removing Whitespace:
whitespace_string = "  Remove extra spaces  "
stripped_string = whitespace_string.strip()
# Output: 'Remove extra spaces'

# 18. Checking Prefix and Suffix:
starts_with_hello = my_string.startswith("Hello")
ends_with_exclamation = my_string.endswith("!")
# Output: True (for both)

# 19. Character Replacement:
replaced_char = my_string.replace("o", "0")
# Output: 'Hell0, Pyth0n!'

# 20. String Reversal:
reversed_string = my_string[::-1]
# Output: '!nohtyP ,olleH'
```

## **List**


```python
my_list = [1, "apple", 3.14, True, ["nested", "list"], {"key": "value"}, (1, 2, 3)]

# Features:
# 1. It can store different data types (int, str, float, bool, list, dictionary, tuple).
# 2. It can store nested data structures (a list within the list).
# 3. It can store key-value pairs (dictionary).
# 4. It is ordered, meaning it maintains the order of elements as they were added.
# 5. It can store duplicate elements (e.g., the number 1 appears twice).
# 6. It is mutable, so you can change its elements.
# 7. It is indexed and iterable, allowing you to access elements by their position.

# Accessing list elements:
first_element = my_list[0]  # Output: 1
nested_list_element = my_list[4][1]  # Output: "list"

# Modifying list elements:
my_list[2] = 2.71  # Update the value at index 2

# Adding elements to the list:
my_list.append("new_item")  # Appends "new_item" to the end of the list
my_list.insert(1, "banana")  # Inserts "banana" at index 1

# Removing elements from the list:
my_list.remove(True)  # Removes the first occurrence of True
popped_element = my_list.pop(3)  # Removes and returns the element at index 3

# List slicing:
subset = my_list[2:5]  # Retrieves elements at indices 2, 3, and 4

# List length:
length = len(my_list)  # Returns the number of elements in the list

# List concatenation:
another_list = ["grape", "cherry"]
combined_list = my_list + another_list  # Combines both lists

# List sorting:
my_list.sort(key=str)  # Sorts the list in ascending order
reversed_list = list(reversed(my_list))  # Reverses the list

# List comprehension:
squared_numbers = [x**2 for x in my_list if isinstance(x, int)]  # Creates a new list of squared integers

# Checking for an element in the list:
contains_apple = "apple" in my_list  # Returns True

# Counting occurrences of an element:
count_of_apples = my_list.count("apple")  # Returns the number of times "apple" appears

# Extending the list:
my_list.extend(["orange", "kiwi"])  # Adds multiple elements to the end of the list

# Clearing the list:
my_list.clear()  # Removes all elements from the list
```

## **Tuple**

```python
my_tuple = (1, "apple", 3.14, True, (4, 5), {"key": "value"})

# Features:
# 1. It can store different data types (int, str, float, bool, tuple, dictionary).
# 2. It can store nested data structures (a tuple within the tuple).
# 3. It is ordered, meaning it maintains the order of elements as they were added.
# 4. It can store duplicate elements (e.g., the number 1 appears twice).
# 5. It is immutable, so you cannot change its elements after creation.
# 6. It is indexed and iterable, allowing you to access elements by their position.

# Accessing tuple elements:
first_element = my_tuple[0]  # Output: 1
nested_tuple_element = my_tuple[4]  # Output: (4, 5)

# Length of the tuple:
length = len(my_tuple)  # Returns the number of elements in the tuple

# Tuple unpacking:
a, b, c, d, e, f = my_tuple  # Assigns elements to variables

# Count of an element in the tuple:
count_of_1 = my_tuple.count(1)  # Returns the number of times 1 appears in the tuple

# Index of an element in the tuple:
index_of_true = my_tuple.index(True)  # Returns the index of the first occurrence of True

# Concatenating tuples:
another_tuple = (7, "orange")
combined_tuple = my_tuple + another_tuple  # Combines both tuples

# Nested tuple indexing:
second_element_of_nested = my_tuple[4][1]  # Accessing the second element of the nested tuple

# Checking for an element in the tuple:
contains_apple = "apple" in my_tuple  # Returns True

# Tuple repetition:
repeated_tuple = my_tuple * 3  # Repeats the tuple three times
```

## **Set**

```python
my_set = {1, 2, 3, 4, 4, 5}

# Features:
# 1. It can only contain unique elements (duplicates are automatically removed).
# 2. It is unordered, meaning elements have no specific order.
# 3. It is mutable, allowing you to add or remove elements after creation.
# 4. It can perform set operations like union, intersection, and difference.
# 5. It is iterable but does not support indexing because of its unordered nature.

# Adding elements to the set:
my_set.add(6)  # Adds 6 to the set
my_set.update({7, 8, 9})  # Adds multiple elements to the set

# Removing elements from the set:
my_set.remove(1)  # Removes the element 1 (raises an error if not found)
my_set.discard(10)  # Removes the element 10 if it exists (no error if not found)

# Set operations:
another_set = {3, 4, 5, 10}
union_set = my_set.union(another_set)  # Combines elements from both sets
intersection_set = my_set.intersection(another_set)  # Retrieves common elements
difference_set = my_set.difference(another_set)  # Retrieves elements in my_set but not in another_set

# Checking for an element in the set:
contains_3 = 3 in my_set  # Returns True

# Length of the set:
set_length = len(my_set)  # Returns the number of elements in the set

# Iterating through the set:
for element in my_set:
    print(element)

# Clearing the set:
my_set.clear()  # Removes all elements from the set
```

## **Dictionary**

```python
my_dict = {
    "name": "John Doe",
    "age": 30,
    "is_student": False,
    "hobbies": ["reading", "hiking"],
    "address": {
        "street": "123 Main St",
        "city": "Anytown"
    }
}

# Features:
# 1. It stores key-value pairs.
# 2. Keys are unique and immutable (usually strings but can be other immutable types).
# 3. Values can be of various data types (str, int, bool, list, nested dictionary).
# 4. It is ordered (as of Python 3.7+), meaning it maintains the order of key-value pairs.
# 5. It is mutable, allowing you to add, update, or remove key-value pairs.
# 6. You can access values using their keys.
# 7. You can iterate through keys, values, or key-value pairs.
# 8. You can check for the existence of a key in the dictionary.
# 9. You can find the length (number of key-value pairs) of the dictionary.

# Accessing dictionary values:
name = my_dict["name"]  # Accessing the value associated with the "name" key
address_street = my_dict["address"]["street"]  # Accessing a nested dictionary value

# Modifying dictionary values:
my_dict["age"] = 31  # Update the value associated with the "age" key

# Adding key-value pairs to the dictionary:
my_dict["email"] = "john@example.com"
my_dict["hobbies"].append("swimming")

# Removing key-value pairs from the dictionary:
del my_dict["is_student"]
removed_hobby = my_dict["hobbies"].pop(0)

# Dictionary keys and values:
keys = my_dict.keys()  # Returns a list of keys
values = my_dict.values()  # Returns a list of values
key_value_pairs = my_dict.items()  # Returns a list of (key, value) tuples

# Iterating through the dictionary:
for key in my_dict:
    value = my_dict[key]
    print(f"{key}: {value}")

# Checking for the existence of a key:
contains_address = "address" in my_dict  # Returns True

# Length of the dictionary:
dict_length = len(my_dict)  # Returns the number of key-value pairs in the dictionary

# Clearing the dictionary:
my_dict.clear()  # Removes all key-value pairs from the dictionary
```

Factorial of a Number

Factorial of a number is the product of all the positive integers less than or equal to the given number

Factorial of 5 = 5 x 4 x 3 x 2 x 1 = 120

```python
number=int(input("Enter the number to find the factorial: "))

factorial=1

if number==0:
  print("factorial of 0 is 1")
else:
  for i in range(1, number+1):
    factorial=factorial*i
  print("the factorial of",number, "is", factorial)
```

```python
def factorial_value(num):

  factorial = 1

  if num == 0:
    return factorial

  else:
    for i in range(1, num+1):
      factorial = factorial*i
    return factorial
```

```python
print(factorial_value(10))
print(factorial_value(5))
print(factorial_value(2))
```
